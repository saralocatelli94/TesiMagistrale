```python
RR_intervals=df['rr']
time_rr=df['time']
```

```python
plt.figure(figsize=(10,3))

plt.plot(time_rr,RR_intervals)
#plt.figure(figsize=(10,3))
#plt.plot(time_apn,apn)
```

```python
avg=statistics.mean(RR_intervals)
RR=[]
timeRR=[]
i=0
while (i<len(RR_intervals)-1):
    somma=RR_intervals[i]+RR_intervals[i+1]
    diff=abs(somma-avg)
    if(abs(diff)>abs(avg-RR_intervals[i])):
        RR.append(RR_intervals[i])
        timeRR.append(time_rr[i])
        i=i+2
    else:
        RR.append(somma)
        timeRR.append(time_rr[i])
        i=i+1
```

```python
window=20
for i in range(0,len(RR_intervals)):
    #print(i)
    if(i>= window):
        start=i-window
    else:
        start=0
    if(i<=len(RR_intervals)-window):
        stop=len(RR_intervals)-window
    else:
        stop=len(RR_intervals)-1
    avg=statistics.mean(RR_intervals[start:stop])
    if(RR_intervals[i]>=0.8*avg):
        RR_intervals[i]=0.8*avg
    
```

```python
window=40
for i in range(0,len(RR_intervals)):
    if(i>=window):
        start=i-window
    else:
        start=0
    if(i<=len(RR_intervals)-window):
        stop=i+window
    else:
        stop=len(RR_intervals)-1
    avg=statistics.mean(RR_intervals[start:stop])
    if(RR_intervals[i]>=0.2*avg or RR_intervals[i]<=-0.2*avg):
        RR_intervals[i]=avg
```
