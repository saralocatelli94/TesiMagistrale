## Definizione di alcune funzioni

#### Funzione che aggiunge una colonna al dataset indicante le durate delle varie fasi

```python
def calcola_durata(dataframe,y,time_ultima_posizione):
    lista_durate=[]
    for i in range(0,len(dataframe)):
        y.append((datetime.timedelta(seconds=dataframe['time'][i])))
        y[i]=datetime.datetime.strptime(str(y[i]),'%H:%M:%S')
        if(i==len(dataframe)-1):
            lista_durate.append(time_ultima_posizione-dataframe['time'][i])
        if(i<=len(dataframe)-2):
            lista_durate.append(dataframe['time'][i+1]-dataframe['time'][i])
    dataframe['durata']=lista_durate
    return dataframe
```

#### Funzione che crea un nuovo dataset contenente la durata totale delle varie fasi/posizioni

```python
def crea_dataset_cumulativo(dataframe, nome_col1, lista_valori_col1):
    df2=pd.DataFrame(columns=[nome_col1,'TempoTotale(s)','TempoTotale'])
    df2[nome_col1]=lista_valori_col1
    tot=[]
    time=[]
    header=list(dataframe.columns.values)
    for i in range(0,len(lista_valori_col1)):
        tot.append(dataframe.loc[dataframe[header[1]] == lista_valori_col1[i], 'durata'].sum())
       # tot.append(df.loc[col1 == i, 'durata'].sum())
        time.append(str(datetime.timedelta(seconds=tot[i])))
    df2['TempoTotale(s)']=tot
    df2['TempoTotale']=time
    return df2
```
